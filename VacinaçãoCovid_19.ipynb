```python
# Total de Vacinas x Fabricante
#fabricante = df.groupby(['vacina_fabricante_nome']).agg({'document_id': 'count'})
fabricante = df['vacina_fabricante_nome'].value_counts()
display(fabricante)
```

```python
fabricante.index
```

```python
fabricante.values
```

```python
#Gráfico Pizza - Fabricante
px.pie(fabricante, names=fabricante.index, values=fabricante.values)
```

```python
# Quantidade de vacinados por município
df.groupby(['estabelecimento_municipio_nome']).agg({'document_id': 'count'})
```

```python
# % de Vacinas por Sexo
df.paciente_enumSexoBiologico.value_counts(normalize=True)
```

```python
# Quantidade de vacinados por Sexo
df_sexo = df.groupby(['paciente_enumSexoBiologico']).agg({'document_id': 'count'}).reset_index()
display(df_sexo)
```

```python
# Gráfico de barras - Sexo
px.bar(df_sexo, x='paciente_enumSexoBiologico', y='document_id', color='paciente_enumSexoBiologico', barmode='group')
```

```python
# Quantidade de vacinados por município
df_mun_idade_vacina = df.groupby(['estabelecimento_municipio_nome','paciente_idade','vacina_fabricante_nome'], sort=False).count().reset_index()
display(df_mun_idade_vacina)
```
